```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
coronaData=pd.read_csv("covid19_Confirmed_dataset.csv")
coronaData.head(10)
```

```python
df=coronaData.drop(["Lat","Long"],axis=1)  #axis=0 is row and axis=1 is column
```

```python
df.head(10)
```

```python
corona_agg=df.groupby("Country/Region").sum()
```

```python
corona_agg.head()
```

```python
corona_agg.loc["India"].diff().plot()
```

```python
countries=list(corona_agg.index)
```

```python
countries
```

```python
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(corona_agg.loc[c].diff().max())
corona_agg["max_infection_rate"]=max_infection_rates
corona_agg.head()
```

```python
df=pd.DataFrame(corona_agg["max_infection_rate"])
df.head()
```

```python
happy_data=pd.read_csv("worldwide_happiness_report.csv")
happy_data.head()
```

```python
happy_data.drop(["Overall rank","Score","Generosity","Perceptions of corruption"],axis=1,inplace=True)
happy_data.set_index("Country or region",inplace=True)
happy_data.head()
```

```python
data=df.join(happy_data,how="inner")
```
